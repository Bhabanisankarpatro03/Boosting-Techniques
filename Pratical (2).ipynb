22.  Train a Gradient Boosting Regressor and plot learning curves.

```python
# Import necessary libraries
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split, learning_curve
import numpy as np
import matplotlib.pyplot as plt

# Generate a synthetic regression dataset
X, y = make_regression(n_samples=1000, n_features=20, noise=0.1, random_state=42)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the Gradient Boosting Regressor
# You can adjust hyperparameters like n_estimators, learning_rate, etc.
gb_regressor = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, random_state=42)

# Compute learning curves
train_sizes, train_scores, test_scores = learning_curve(
    gb_regressor, X_train, y_train, cv=5, scoring='neg_mean_squared_error',
    train_sizes=np.linspace(0.1, 1.0, 10), random_state=42)

# Convert negative MSE to positive
train_scores = -train_scores
test_scores = -test_scores

# Calculate mean and standard deviation of training and test scores
train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)
test_scores_std = np.std(test_scores, axis=1)

# Plot learning curves
plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_scores_mean, label="Training score", color="blue")
plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                 train_scores_mean + train_scores_std, alpha=0.1, color="blue")
plt.plot(train_sizes, test_scores_mean, label="Cross-validation score", color="red")
plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                 test_scores_mean + test_scores_std, alpha=0.1, color="red")

# Add labels and title
plt.xlabel("Training Set Size")
plt.ylabel("Mean Squared Error (MSE)")
plt.title("Learning Curves for Gradient Boosting Regressor")
plt.legend(loc="best")
plt.grid()
plt.show()
```

25.  Train an AdaBoost Classifier with different numbers of estimators and compare accuracy.

```python
# Import necessary libraries
from sklearn.ensemble import AdaBoostClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Load the Breast Cancer dataset
data = load_breast_cancer()
X = data.data  # Features
y = data.target  # Labels (0 = malignant, 1 = benign)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define a list of different numbers of estimators
n_estimators_list = [10, 50, 100, 200, 300]

# Initialize lists to store accuracy scores
train_accuracy = []
test_accuracy = []

# Train AdaBoost Classifier with different numbers of estimators
for n_estimators in n_estimators_list:
    # Initialize the AdaBoost Classifier
    adaboost = AdaBoostClassifier(n_estimators=n_estimators, random_state=42)

    # Train the classifier on the training data
    adaboost.fit(X_train, y_train)

    # Make predictions on the training and test data
    y_train_pred = adaboost.predict(X_train)
    y_test_pred = adaboost.predict(X_test)

    # Calculate accuracy for training and test data
    train_accuracy.append(accuracy_score(y_train, y_train_pred))
    test_accuracy.append(accuracy_score(y_test, y_test_pred))

    # Print accuracy for the current number of estimators
    print(f"Number of Estimators: {n_estimators}")
    print(f"Training Accuracy: {train_accuracy[-1] * 100:.2f}%")
    print(f"Test Accuracy: {test_accuracy[-1] * 100:.2f}%")
    print("-----------------------------")

# Plot the accuracy vs. number of estimators
plt.figure(figsize=(10, 6))
plt.plot(n_estimators_list, train_accuracy, label="Training Accuracy", marker='o')
plt.plot(n_estimators_list, test_accuracy, label="Test Accuracy", marker='o')
plt.xlabel("Number of Estimators")
plt.ylabel("Accuracy")
plt.title("Accuracy vs. Number of Estimators for AdaBoost Classifier")
plt.legend()
plt.grid()
plt.show()
```

28.  Train a CatBoost Classifier on an imbalanced dataset and compare performance with class weighting.

```python
# Import necessary libraries
from catboost import CatBoostClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Generate an imbalanced synthetic dataset
X, y = make_classification(n_samples=1000, n_features=20, n_informative=15,
                           n_redundant=5, n_clusters_per_class=1, weights=[0.95, 0.05],
                           random_state=42)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the CatBoost Classifier without class weighting
catboost_unweighted = CatBoostClassifier(random_state=42, verbose=0)

# Train the classifier on the training data
catboost_unweighted.fit(X_train, y_train)

# Make predictions on the test data
y_pred_unweighted = catboost_unweighted.predict(X_test)

# Evaluate the unweighted model
print("Unweighted Model Performance:")
print(classification_report(y_test, y_pred_unweighted))
print("Confusion Matrix:")
disp = ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred_unweighted),
                              display_labels=["Class 0", "Class 1"])
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix (Unweighted)")
plt.show()

# Initialize the CatBoost Classifier with class weighting
# Calculate class weights based on the inverse of class frequencies
class_weights = {0: 1, 1: len(y_train[y_train == 0]) / len(y_train[y_train == 1])}
catboost_weighted = CatBoostClassifier(random_state=42, verbose=0, class_weights=class_weights)

# Train the classifier on the training data
catboost_weighted.fit(X_train, y_train)

# Make predictions on the test data
y_pred_weighted = catboost_weighted.predict(X_test)

# Evaluate the weighted model
print("Weighted Model Performance:")
print(classification_report(y_test, y_pred_weighted))
print("Confusion Matrix:")
disp = ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred_weighted),
                              display_labels=["Class 0", "Class 1"])
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix (Weighted)")
plt.show()
```

29.  Train an AdaBoost Classifier and analyze the effect of different learning rates.

```python
# Import necessary libraries
from sklearn.ensemble import AdaBoostClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Load the Breast Cancer dataset
data = load_breast_cancer()
X = data.data  # Features
y = data.target  # Labels (0 = malignant, 1 = benign)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define a list of learning rates to analyze
learning_rates = [0.01, 0.05, 0.1, 0.2, 0.5, 1.0]

# Initialize lists to store accuracy scores
train_accuracy = []
test_accuracy = []

# Train AdaBoost Classifier with different learning rates
for lr in learning_rates:
    # Initialize the AdaBoost Classifier
    adaboost = AdaBoostClassifier(n_estimators=50, learning_rate=lr, random_state=42)

    # Train the classifier on the training data
    adaboost.fit(X_train, y_train)

    # Make predictions on the training and test data
    y_train_pred = adaboost.predict(X_train)
    y_test_pred = adaboost.predict(X_test)

    # Calculate accuracy for training and test data
    train_accuracy.append(accuracy_score(y_train, y_train_pred))
    test_accuracy.append(accuracy_score(y_test, y_test_pred))

    # Print accuracy for the current learning rate
    print(f"Learning Rate: {lr}")
    print(f"Training Accuracy: {train_accuracy[-1] * 100:.2f}%")
    print(f"Test Accuracy: {test_accuracy[-1] * 100:.2f}%")
    print("-----------------------------")

# Plot the accuracy vs. learning rate
plt.figure(figsize=(10, 6))
plt.plot(learning_rates, train_accuracy, label="Training Accuracy", marker='o')
plt.plot(learning_rates, test_accuracy, label="Test Accuracy", marker='o')
plt.xlabel("Learning Rate")
plt.ylabel("Accuracy")
plt.title("Accuracy vs. Learning Rate for AdaBoost Classifier")
plt.legend()
plt.grid()
plt.show()
```
